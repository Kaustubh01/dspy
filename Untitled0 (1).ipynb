```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
import warnings
warnings.filterwarnings("ignore")
```

```python
# Reading the dataset from the csv file
# separator is a vertical line, as seen in the dataset
data = pd.read_csv("iris.data.csv")

# Printing the shape of the dataset
print(data.shape)
```

```python
data.head()
```

```python

X = data.iloc[:,:-1]
y = data.iloc[:,-1]
print("Shape of X is %s and shape of y is %s"%(X.shape,y.shape))
```

```python
total_classes = y.nunique()
print("Number of unique species in dataset are: ",total_classes)
```

```python
distribution = y.value_counts()
print(distribution)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, y, test_size=0.25, random_state=28)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_curve, auc
import matplotlib.pyplot as plt

# Creating AdaBoost classifier model
adb = AdaBoostClassifier()
adb_model = adb.fit(X_train, Y_train)

# Making predictions
Y_pred = adb_model.predict(X_test)
Y_proba = adb_model.predict_proba(X_test)[:, 1]  # Probability estimates for the positive class

# Calculating metrics
accuracy = accuracy_score(Y_test, Y_pred)

# Get the classification report
report = classification_report(Y_test, Y_pred, output_dict=True)

# Check for the existence of classes '0' and '1'
precision = report['1']['precision'] if '1' in report else None
recall = report['1']['recall'] if '1' in report else None
f1 = report['1']['f1-score'] if '1' in report else None

conf_matrix = confusion_matrix(Y_test, Y_pred)

# Print the metrics
print(f"Accuracy: {accuracy}")
print(f"Precision (class 1): {precision}")
print(f"Recall (class 1): {recall}")
print(f"F1-score (class 1): {f1}")
print("Confusion Matrix:\n", conf_matrix)
```
